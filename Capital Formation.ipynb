```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("API_NE.GDI.TOTL.ZS_DS2_en_csv_v2_423968.csv")
```

```python
data.set_index("Country Name", inplace = True)
```

```python
filtro = ["Belgium","Czech Republic","France","Germany","Italy","Netherlands","Poland","Portugal","Romania",
          "Spain","Sweden","United Kingdom","Norway","Turkey",
          "Kazakhstan","Russian Federation","Ukraine","Uzbekistan","United States","Canada","Argentina","Brazil",
          "Chile","Colombia","Mexico","Venezuela, Bolivarian Republic of","China","India","Indonesia","Japan","Malaysia",
          "Korea, Republic of","Taiwan, Province of China","Thailand","Australia","New Zealand","Algeria","Egypt","Nigeria",
          "South Africa","Iran, Islamic Republic of","Kuwait","Saudi Arabia","United Arab Emirates"]
```

```python
data2 = data.loc[filtro]
```

```python
data2.head(44)
```

```python
data2 = data2.reset_index()
```

```python
data_tidy = pd.melt(data2, id_vars = ['Country Code', 'Country Name'], var_name = ['year'], value_name = 'cap_for')
```

```python
data_tidy = data_tidy.set_index(['Country Name', 'Country Code']).sort_index()
```

```python
data_tidy.to_csv('capital_formation.csv', sep = ',')
```
